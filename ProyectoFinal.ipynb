## Llenado de datos categricos perdidos
Aqu tenemos el dilema de que los datos vacos tienen un significado en el reto, por ejemplo un NA: No alley access, para este problema vamos a crear un valor constante "None". 

Para esto primero debimos revisar rigurosamente el contenido de los datos que contienen vacos y ver si se trata de datos cuyo vaco significa algo y si es as, reemplazar con el "None".

```python
#Reemplazamos los vacos, en las columnas cuyo vaco tiene significado: 
for column in ['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish',
               'GarageQual','GarageCond','PoolQC','Fence','MiscFeature']:
    data2[column] = data2[column].fillna("None")
#Reemplazamos los vacos, en las columnas cuyo vaco no tiene significado:
for column in ['MSZoning','Utilities','Exterior1st','Exterior2nd','MasVnrType','Electrical','KitchenQual','Functional','SaleType']:
    data2[column] = data2[column].fillna(data2[column].mode()[0])
```

##Llenado de datos numricos vacos


```python
#Hacemos una copia de data2
data3 = data2.copy()

#Creamos una funcin que llene los datos numricos vacos, usando KNeighborsRegressor
def value_knn(df, na_target):
    df = df.copy()
    #Seleccionamos nicamente los datos numpericos del df
    numeric_df = df.select_dtypes(np.number)
    #Estas son los valores de las columnas que contienen datos numricos que no tienen datos vacos
    non_na_columns = numeric_df.loc[: ,numeric_df.isna().sum() == 0].columns
    #bsicamente vamos a usar esta informacin para rellenar los vacos
    #y_train tiene todos los valores de las columnas dentro de na_target que no son vacos
    y_train = numeric_df.loc[numeric_df[na_target].isna() == False, na_target]
    #x_train contiene todos los valores de las columnas de non_na_columns (el resto de los datos) que no son vacos 
    X_train = numeric_df.loc[numeric_df[na_target].isna() == False, non_na_columns]
    #X_test contienen todos los valores de las columnas que contienen datos vacos
    X_test = numeric_df.loc[numeric_df[na_target].isna() == True, non_na_columns]
    #Creamos el regressor
    knn = KNeighborsRegressor()
    #Aplicamos el fit
    knn.fit(X_train, y_train)
    #Hacemos la prediccin 
    y_pred = knn.predict(X_test)
    #finalmente y_pred contiene el valor con el que vamos a llenar los vacos
    df.loc[df[na_target].isna() == True, na_target] = y_pred
    
    return df

#Usamos la funcin anterior para rellenar los datos vacos en todas las columnas:
for column in ['LotFrontage','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageYrBlt',
               'GarageCars','GarageArea']:
    data3 = value_knn(data3, column) #Definimos data3 
data4 = data3.copy() #hacemos una copy por si las moscas 
```

##Transformacin de Caractersticas
Buscamos transformar nuestros datos para que tengan una distribucin normal, de esta manera esperamos que el modelo se desenvuelva mucho mejor. Lo que buscamos es volver ms normal (simtrica) la distribucin de los datos que puede estar inclinada a la derecha o a la izquierda.


```python
#observemos la inclinacin (skew) de las columnas. Si este valor es negativo, significa que la mayora de los datos se encuentran 
#a la  izquierda, si es cero significa que es perfectamente simtrico y si es positivo, significa que est inclinada hacia la derecha.
pd.DataFrame(scipy.stats.skew(data4.select_dtypes(np.number)))
#Podemos ver cmo est la inclinacin de las columnas que contienen valores numricos 
```

```python
data4["SqFtPerRoom"] = data4["GrLivArea"] / (data4["TotRmsAbvGrd"] +
                                                       data4["FullBath"] +
                                                       data4["HalfBath"] +
                                                       data4["KitchenAbvGr"])

data4['Total_Home_Quality'] = data4['OverallQual'] + data4['OverallCond']

data4['Total_Bathrooms'] = (data4['FullBath'] + (0.5 * data4['HalfBath']) +
                               data4['BsmtFullBath'] + (0.5 * data4['BsmtHalfBath']))

data4["HighQualSF"] = data4["1stFlrSF"] + data4["2ndFlrSF"]

data5 = data4.copy()
```

```python
#observemos la inclinacin (skew) de las columnas. Si este valor es negativo, significa que la mayora de los datos se encuentran 
#a la  izquierda, si es cero significa que es perfectamente simtrico y si es positivo, significa que est inclinada hacia la derecha.
#Podemos ver cmo est la inclinacin de las columnas que contienen valores numricos 
skew_df = pd.DataFrame(data5.select_dtypes(np.number).columns, columns=['Feature']) #Creamos un dataframe para almacenar las inclinaciones 

skew_df['Skew'] = skew_df['Feature'].apply(lambda feature: scipy.stats.skew(data5[feature])) #sacamos la inclinacin de cada una de las columnas y la almacenamos

skew_df['Absolute Skew'] = skew_df['Skew'].apply(abs) #Como tenemos algunos valores negativos aplicamos valor absoluto y almacenamos en otra columna 

skew_df['Skewed'] = skew_df['Absolute Skew'].apply(lambda x: True if x >= 0.5 else False) #Ahora creamos otra columna para verificar si est inclinado basado en el criterio de +0.5 y almacenamos

for column in skew_df.query("Skewed == True")['Feature'].values:
    data5[column] = np.log1p(data5[column])

data4['MoSold'] = (-np.cos(0.5236 * data5['MoSold']))

data6 = data5.copy()
```
